# **Answer 1a**

In predictive inference, data science, and machine learning we generally care about model performance, specifically on "out-of-sample" predictions. Our goal is "simple" but difficult to achieve: we want "good" results from our models. In this case, a "good result" is a model that **generalizes** well in its prediction on data *it has not been trained on*. Fundamentally, that's the goal of predictive inference—to understand and forecast a future/current world based on historical data, even if we don't *exactly* understand why we are predicting so well.

In causal inference (analysis), this is not the case. Practitioners of causal inference care much more about identifying cause. Specfically, practitioners look to identify the average treatment effects on those who received some sort of "treatment" (that could be those who were provided a special study guide) compared to those who did not receive that treatment. This is critical for, say, social scientists who want to make bold claims such as "minimum wage *causes* X to happen".

To achieve this, causal inference relies on the understanding that, in order to "determine" a true cause of some treatment effect, the differences in groups being assessed must be, for all intents and purposes, equal. That is, we must account for bias in our control and treatement groups.

For example, if we return to the our assessment of whether certain study programs are effective, we will want to account for "motivation" between Treatment and Control groups. Generally, we perform Randomized Control Trials such that individuals who end up in control and treatment groups would be statistically (on average) the same. That way, any statistically signficant differences in Treatment versus Control group results should indicate a "true" causal effect.

Propensity scores are a fundamental concept in causal inference and are used to **estimate the effect of a treatment, policy, or other intervention by accounting for covariates (variables) that predict receiving the treatment.**

[In class two we were] trying to determine the effect of a study program on students' final grades. However, students who participate in the study program might be different from those who don't in various ways - they might be more motivated, have more free time, or have better prior academic records. These differences can create **bias** when estimating the program's effect.

This is where propensity scores come into play. **A propensity score for a given student would represent the probability that they participate in the study program, given their observed characteristics (like motivation level, amount of free time, prior academic records, etc.).** These scores allow us to balance or match students who participated in the program with those who didn't, in a way that makes the two groups as similar as possible. **Therefore, any remaining differences in outcomes can be more confidently attributed to the program itself, rather than pre-existing differences among the students.**
