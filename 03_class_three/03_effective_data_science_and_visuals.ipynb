### **Comma Delimited Values (CSVs)**

**What is a delimited file?**

A delimited file is a plain text file where individual data values are separated by specific characters, called delimiters (e.g., commas, tabs).

**Common Delimiters**

- Comma: `,`
  - Used in CSV - Comma-Separated Values
- Tab: `\t`
  - Used in TSV - Tab-Separated Values
- Semicolon: `;`
  - Common in European data files due to the comma being used as a decimal point

**Why comma delimited files?**
- Easy to write preformated tabular data.
- Each line = one row; values separated by commas.

**Uses**:

- Import/export data from databases.
- Data interchange between different applications.
- Easily read by spreadsheet programs.

### **Accessing Files: Relative Versus Absolute Paths**

**Relative Path**:

Defined based on the current location you are working from called your `current directory` or `working directory`!

Example: `./images/photo.jpg` (from the working directory (`./`), go to the `images/` folder and access `photo.jpg`).

**Absolute Path**:

Specifies the full path from the root to the file or directory.

Example: `/home/user/images/photo.jpg` (starts from the root directory (`/home/` and gives the complete path).

 **Key Point**:

- Relative paths are shorter and more convenient, use them when you can!
- Use absolute paths when accessing your file via its relative path isn't working, or when you want to be certain the file is read.

### Displaying Your Data

`Altair`: A Python visualization library that allows for easy creation of interactive visualizations with a simple syntax.

```python
import altair as alt

# Visualization 1: Histogram of IMDB Scores
alt.Chart(movies).mark_bar().encode(
    x=alt.X("imdb_score:Q", bin=alt.Bin(maxbins=30), title="IMDB Score"),
    y='count()',
    tooltip=['count()']
).properties(
    title="Distribution of IMDB Scores",
    width=300,
    height=200
)

```

```python
# Visualization 2: Top Directors
top_directors = movies['director_name'].value_counts().reset_index().head(10)
top_directors.columns = ['Director', 'Number of Movies']

bar_directors = alt.Chart(top_directors).mark_bar().encode(
    x='Number of Movies:Q',
    y=alt.Y('Director:O', sort='-x'),
    tooltip=['Director', 'Number of Movies']
).properties(
    title="Top 10 Directors by Number of Movies",
    width=300,
    height=200
)

bar_directors
```

```python
# Visualization 3: Gross vs Budget
scatter_gross_budget = alt.Chart(movies.sample(1000)).mark_circle().encode(
    x=alt.X('budget:Q', title="Budget"),
    y=alt.Y('gross:Q', title="Gross Revenue"),
    color='imdb_score:Q',
    tooltip=['movie_title', 'budget', 'gross', 'imdb_score']
).properties(
    title="Gross Revenue vs. Budget",
    width=300,
    height=200
)

scatter_gross_budget
```

```python
# Visualization 4: Most Common Movie Genres
df_genres = movies['genres'].str.split('|', expand=True).stack().reset_index(drop=True).value_counts().reset_index().head(10)
df_genres.columns = ['Genre', 'Count']

bar_genres = alt.Chart(df_genres).mark_bar().encode(
    x='Count:Q',
    y=alt.Y('Genre:O', sort='-x'),
    tooltip=['Genre', 'Count']
).properties(
    title="Top 10 Most Common Movie Genres",
    width=300,
    height=200
)

bar_genres
```

```python
# Visualization 5: Trend of Average IMDB scores over the years
line_trend = alt.Chart(movies).mark_line().encode(
    x=alt.X('title_year:O', title="Year"),
    y=alt.Y('mean(imdb_score):Q', title="Average IMDB Score"),
    tooltip=['title_year', 'mean(imdb_score)']
).properties(
    title="Trend of Average IMDB Scores Over the Years",
    width=300,
    height=200
)

line_trend
```
