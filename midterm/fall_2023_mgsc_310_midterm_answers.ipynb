# Question 2

That chart was confusing! It was too small. And also that data didn't represent today. Let's try something else.

- a. Load the `Fund` data from the ISLP package [here](https://islp.readthedocs.io/en/latest/datasets/Fund.html). Assign it to the variable `fund`. This is information about the managers I pay to trade my money around. I suspect that some are making systemically bad trades. I want you to identify them. (1pt)

- b. Take the mean of every fund manager. *sort* the *values*. Assign this result to the variable `avg_vals`. run `describe` on `avg_vals`. Assign it to the variable `dist` and display its results. (2pts)


- c. The `Empirical Rule`, or the `68-95-99.7 Rule`, says that around 68% of results should lie within 1 standard deviation of the mean, around 95% of scores within 2 standard deviations of the mean, and around 99.7% of scores should lie within 3 standard deviations of the mean, if we assume a normal distribution.

  Assign the `mean` from `dist` to variable `mu`, and assign standard deviation from `dist` to the variable `sig`. Calculate what three standard deviations below the mean should be, assign it to the variable `three_sig_below`. (3pts)

- d. Filter `avg_vals` down to only include data that is less than `three_sig_below`. Display who the underperformers are by printing out the filtered `avg_vals` results. These are the managers who are systemically underperforming... They will be dealt with. (2pts)

````python
# a
fund = load_data("Fund")

# b
avg_vals = fund.mean().sort_values()

dist = avg_vals.describe()

# c
mu = dist['mean']
sig = dist['std']
three_sig_below = mu - (3*sig)

three_sig_below
````

````python
# d: systemic underperformers
avg_vals[avg_vals < three_sig_below]
````

- e. Run a [LogisticRegression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html) on your `X_train` and `y_train` data. Then predict on your `X_test` data. (2pts)

````python
# a

credit = load_data('Credit')

credit['is_high'] = (credit['Rating'] > 300).astype(int)

# another method
# credit['is_high'] = credit['Rating'].map(lambda x: 1 if x > 500 else 0)

credit.shape
````

````python
# b / c
mapping = {"No": 0, "Yes":1}
credit['is_student'] = credit['Student'].map(mapping)
credit['is_married'] = credit['Married'].map(mapping)
````

````python
# d.
X = credit[['Income', 'Limit', 'Cards', 'Age', 'Education', 'Balance', 'is_student', 'is_married']]
y = credit['is_high']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25,
                                                    random_state=643, shuffle=True)
````

````python
#e.
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
````

````python
y_pred = log_reg.predict(X_test)
````

````python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
````

- The results are pretty great! We have 19 True Negatives and 37 True Positives. predicted our True Positives and True Negatives in our out of sample prediction. That means the model generalizes very well. We predicted 3 False Negatives and 1 False Positive.

````python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test, y_pred)
````
