```python
# In this code, we are plotting out a funny barplot that counts the number of
# letters in each word in a sentence

import matplotlib.pyplot as plt # <- This is our plotting library!

sentence = "Ran my first Python code in MGSC 310 and all I got was this measly python plot. Womp."

words = sentence.split()

num_letters = [len(word) for word in words]

plt.figure(figsize=(12, 6))
plt.bar(words, num_letters, color='teal')

plt.xticks(rotation=45, ha='right')

plt.xlabel('Words')
plt.ylabel('Number of Letters')
plt.title('A "Measly" Python Plot')

plt.tight_layout()
plt.show()
```
